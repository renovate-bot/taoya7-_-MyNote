### 自定义样式

1. 颜色
2. 线宽
3. 样式
    - `-` 实线
    - `--` 虚线
    - `:` 点线
    - `-.` 短线

```python
import matplotlib.pyplot as plt

#定义x、y轴数据

x_data = ['2015','2016','2017','2018','2019','2020']
y_data1 = ['5000','6000','8000','15000','25000','38000']
y_data2 = ['3000','6000','9000','15000','25000','55000']

#调用方式2
# plt.plot(x_data,y_data1, x_data,y_data2)

plt.plot(x_data,y_data1, color="red", linewidth=3, linestyle="--")
plt.plot(x_data,y_data2, color="blue", linewidth=5, linestyle="-.")

plt.show() #show()方法 显示图形
```
